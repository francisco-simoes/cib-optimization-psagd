# 2D illustration of how random points on a ball can be generated

(This will be used in pSAGD when generating proposals).

```python
def random_sphere_points(num_points, dimension, radius=1):
    random_ball_points = torch.randn(dimension, num_points)
    random_directions = random_ball_points / torch.norm(random_ball_points, dim=0)
    return radius * random_directions
```

```python
points = random_sphere_points(100, 2, 4)
```
