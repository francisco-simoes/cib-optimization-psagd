```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Probability decay curves for different temperatures and cooling rates

```python
# Typical loss change values for simulated annealing proposals in our experiments
changes = [1e-1, 1e0]

# Define the range of temperature and cooling rate values
T_values = [1, 10, 100]
r_values = [0.9, 0.99, 0.999]

# Define the range of indices
indices = np.arange(1000)

# Create a plot
plt.figure()

# Loop over the values of T and plot the corresponding curves
for change in changes:
    for T in T_values:
        for r in r_values:
            y_values = np.array([np.exp(-change/(T*r**i)) for i in indices])
            plt.plot(indices, y_values, label=f'T={T} r={r}')
    
    # Add labels and legend
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.title(f'Exponential decay of acceptance probabilities for different temperature values and change = {change}')
    plt.show()
```

### Conclusion

r=0.99 is a reasonable cooling rate for different values of the temperature, and for a typical loss changes around 0.1 or 1.0.
Furthermore, temperatures of 10 and 100 lead to probabilities which decay to zero early enough, if one intends to run the learning algorithm for at most 1000 iterations.
